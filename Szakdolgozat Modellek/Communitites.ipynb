```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def plot_comparison(df, dates, ncols=5, figsize=(20,8)):
    # Optional: choose a built-in style
    plt.style.use('ggplot')

    nplots = len(dates)
    nrows  = int(np.ceil(nplots/ncols))
    fig, axes = plt.subplots(
        nrows, ncols, figsize=figsize,
        sharey=True, squeeze=False
    )
    axes = axes.flatten()

    for ax, dt in zip(axes, dates):
        day_mask = df['ts_orig'].dt.date == pd.to_datetime(dt).date()
        sub = df.loc[day_mask]
        if sub.empty:
            ax.set_title(f"{dt}\n(no data)")
            ax.axis('off')
            continue

        ax.plot(
            sub['ts_orig'], sub['production'],
            marker='o', ms=4, lw=1, label='Actual'
        )
        ax.plot(
            sub['ts_orig'], sub['Predicted_Production'],
            marker='x', ms=4, lw=1, label='Predicted'
        )
        ax.set_title(str(dt))
        ax.xaxis.set_major_locator(mdates.HourLocator(interval=2))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        ax.tick_params(axis='x', rotation=45)
        ax.grid(True, which='major', linestyle='--', alpha=0.4)

    # turn off any unused axes
    for ax in axes[nplots:]:
        ax.axis('off')

    # shared legend
    handles, labels = axes[0].get_legend_handles_labels()
    fig.legend(handles, labels, loc='upper right', ncol=2)
    fig.tight_layout(rect=[0,0,0.95,1])
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.model_selection import train_test_split
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score,
    accuracy_score
)
from xgboost import XGBClassifier, XGBRegressor

# -----------------------------------------------------------------------------
# 0) (optional) load your data
# -----------------------------------------------------------------------------
# train = pd.read_csv('train.csv')
# test  = pd.read_csv('test.csv')

# -----------------------------------------------------------------------------
# 1) Prepare data & features
# -----------------------------------------------------------------------------
def add_time_features(df):
    df['year']  = df['timestamp'].dt.year
    df['month'] = df['timestamp'].dt.month
    df['day']   = df['timestamp'].dt.day
    # cyclic encoding of month
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
    return df

for df in (train, test):
    # ensure datetime
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df.sort_values('timestamp', inplace=True)
    df = add_time_features(df)
    # keep original for plotting
    df['ts_orig'] = df['timestamp']

# encode categoricals
for col in ['season', 'category']:
    filename = 'scaler_' + col + '.pkl'
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col]  = le.transform(test[col])
    pickle.dump(le, open(filename, 'wb'))

# create binary label: day vs. night
train['is_day'] = (train['production'] > 0).astype(int)

# -----------------------------------------------------------------------------
# 2) Two-stage modelling
# -----------------------------------------------------------------------------
# 2a) Classifier: is there production?
clf_features = [
    c for c in train.columns
    if c not in ('production','is_day','ts_orig','timestamp')
]
Xc, yc = train[clf_features], train['is_day']
Xc_train, Xc_val, yc_train, yc_val = train_test_split(
    Xc, yc, test_size=0.2, random_state=42, stratify=yc
)
print(clf_features)
pickle.dump(clf_features, open('clf_features.json', 'wb'))

clf = XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)
clf.fit(Xc_train, yc_train)
yc_val_pred = clf.predict(Xc_val)
print(f"Classifier accuracy: {accuracy_score(yc_val, yc_val_pred):.3f}")
pickle.dump(clf, open('clf.sav', 'wb'))

# 2b) Regressor: only on daytime rows
reg_df = train[train['is_day']==1].copy()
reg_features = [
    c for c in reg_df.columns
    if c not in ('production','is_day','ts_orig','timestamp')
]
print(reg_features)
pickle.dump(reg_features, open('reg_features.json', 'wb'))

Xr, yr = reg_df[reg_features], reg_df['production']
Xr_train, Xr_val, yr_train, yr_val = train_test_split(
    Xr, yr, test_size=0.2, random_state=42
)

reg = XGBRegressor(
    objective='reg:squarederror',
    n_estimators=200,
    random_state=42
)
reg.fit(Xr_train, yr_train)
yr_val_pred = reg.predict(Xr_val)
pickle.dump(reg, open('reg.sav', 'wb'))

print("Regressor on daytime data:")
print(f"  RMSE: {np.sqrt(mean_squared_error(yr_val, yr_val_pred)):.3f}")
print(f"   MAE: {mean_absolute_error(yr_val, yr_val_pred):.3f}")
print(f"    R²: {r2_score(yr_val, yr_val_pred):.3f}")

# -----------------------------------------------------------------------------
# 3) Predict on test set
# -----------------------------------------------------------------------------
Xt = test[clf_features].copy()
is_day_pred = clf.predict(Xt)

pred = np.zeros(len(test))
day_idx = np.where(is_day_pred == 1)[0]
if day_idx.size:
    pred[day_idx] = reg.predict(Xt.iloc[day_idx])
test['Predicted_Production'] = pred

# -----------------------------------------------------------------------------
# 4) Plotting: actual vs. predicted
# -----------------------------------------------------------------------------
def plot_comparison(df, dates, ncols=5, figsize=(20,8)):
    plt.style.use('ggplot')  # or 'fivethirtyeight', etc.

    nplots = len(dates)
    nrows  = int(np.ceil(nplots / ncols))
    fig, axes = plt.subplots(
        nrows, ncols,
        figsize=figsize,
        sharey=True, squeeze=False
    )
    axes = axes.flatten()

    for ax, dt in zip(axes, dates):
        mask = df['ts_orig'].dt.date == pd.to_datetime(dt).date()
        sub = df.loc[mask]
        if sub.empty:
            ax.set_title(f"{dt}\n(no data)")
            ax.axis('off')
            continue

        ax.plot(
            sub['ts_orig'], sub['production'],
            marker='o', ms=4, lw=1, label='Actual'
        )
        ax.plot(
            sub['ts_orig'], sub['Predicted_Production'],
            marker='x', ms=4, lw=1, label='Predicted'
        )
        ax.set_title(str(dt))
        ax.xaxis.set_major_locator(mdates.HourLocator(interval=2))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
        ax.tick_params(axis='x', rotation=45)
        ax.grid(True, which='major', linestyle='--', alpha=0.4)

    # turn off extra subplots
    for ax in axes[nplots:]:
        ax.axis('off')

    # shared legend
    handles, labels = axes[0].get_legend_handles_labels()
    fig.legend(handles, labels, loc='upper right', ncol=2)
    fig.tight_layout(rect=[0, 0, 0.95, 1])
    plt.show()

# Example usage: pick dates you want to inspect
dates_to_plot = [
    "2024-05-14", "2024-02-03", "2024-01-08",
    "2024-04-25", "2024-10-25", "2024-07-28"
]
plot_comparison(test, dates_to_plot, ncols=3, figsize=(18,10))

# -----------------------------------------------------------------------------
# 5) (optional) save predictions
# -----------------------------------------------------------------------------
# test.to_csv('test_with_predictions_twostage.csv', index=False)
```
